```python
# Setup and Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import sys

# ‚úÖ Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

# Imports
from utils.visual_utils import plot_multiple_distributions
from utils.model_utils import summarize_simulated_distributions
```

```python
# üé≤ Simulate 4 Different Distributions
np.random.seed(42)
dist_data = {
    "Normal (Œº=0, œÉ=1)": np.random.normal(loc=0, scale=1, size=1000),
    "Skewed (Exponential)": np.random.exponential(scale=1.0, size=1000),
    "Uniform (0, 1)": np.random.uniform(0, 1, size=1000),
    "Bimodal (2 Normals)": np.concatenate([np.random.normal(-2, 0.5, 500),
                                           np.random.normal(2, 0.5, 500)])
}
```

```python
# üìà Plot All Distributions
plot_multiple_distributions(dist_data, bins=30)
```

```python
# üìä Summarize Distribution Properties
summary_df = summarize_simulated_distributions(dist_data)
display(summary_df)
summary_df.to_csv(EXPORT_TABLES / "distribution_summary.csv", index=False)
```
